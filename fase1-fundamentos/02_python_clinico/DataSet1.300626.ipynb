```python
import random
import csv

random.seed(42)

def generar_paciente(id_num):
    edad = random.randint(30, 80)
    sexo = random.choice(["M", "F"])
    peso = round(random.uniform(50, 110), 1)
    altura = round(random.uniform(1.50, 1.85), 2)
    imc = round(peso / (altura ** 2), 1)
    glucosa = random.randint(70, 320)
    hba1c = round(random.uniform(4.5, 11.0), 1)
    sistolica = random.randint(100, 180)
    diastolica = random.randint(60, 110)

    if hba1c >= 6.5 or glucosa >= 126:
        diagnostico = "Diabetes tipo 2"
    elif hba1c >= 5.7 or glucosa >= 100:
        diagnostico = "Prediabetes"
    else:
        diagnostico = "Sin alteracion glucemica"

    return {
        "id": f"P{id_num:03d}",
        "edad": edad,
        "sexo": sexo,
        "peso_kg": peso,
        "altura_m": altura,
        "imc": imc,
        "glucosa_ayuno": glucosa,
        "hba1c": hba1c,
        "presion_sistolica": sistolica,
        "presion_diastolica": diastolica,
        "diagnostico_glucemico": diagnostico
    }

pacientes = [generar_paciente(i) for i in range(1, 51)]

print(f"Pacientes generados: {len(pacientes)}")
print(f"\nPrimer registro:")
for clave, valor in pacientes[0].items():
    print(f"  {clave}: {valor}")

diabeticos = [p for p in pacientes if p["diagnostico_glucemico"] == "Diabetes tipo 2"]
print(f"\nPacientes con Diabetes tipo 2: {len(diabeticos)}")

edades = [p["edad"] for p in pacientes]
print(f"Edad promedio: {sum(edades)/len(edades):.1f} años")
```

```python
def resumen_dataset(pacientes):
    total = len(pacientes)
    print(f"Total de pacientes: {total}")

    mujeres = [p for p in pacientes if p["sexo"] == "F"]
    mujeres_porc = (len(mujeres) / len(pacientes)) * 100
    print(f"Porcentaje de mujeres: {mujeres_porc:.1f}%")

    edades = [p["edad"] for p in pacientes]
    print(f"Edad minima: {min(edades)}")
    print(f"Edad maxima: {max(edades)}")

    diabeticos = [p for p in pacientes if p["diagnostico_glucemico"] == "Diabetes tipo 2"]
    prediabeticos = [p for p in pacientes if p["diagnostico_glucemico"] == "Prediabetes"]
    sin_alteracion = [p for p in pacientes if p["diagnostico_glucemico"] == "Sin alteracion glucemica"]
    print(f"Diabetes tipo 2: {len(diabeticos)}")
    print(f"Prediabetes: {len(prediabeticos)}")
    print(f"Sin alteracion glucemica: {len(sin_alteracion)}")

    alto_riesgo = [p for p in pacientes if p["edad"] > 60 and p["imc"] > 30]
    print(f"Alto riesgo cardiometabolico: {len(alto_riesgo)}")

resumen_dataset(pacientes)
```
