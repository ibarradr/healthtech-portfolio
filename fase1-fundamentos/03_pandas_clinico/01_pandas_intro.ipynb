```python
import pandas as pd

df = pd.read_csv("../02_python_clinico/pacientes_simulados.csv")

print(f"Forma del dataset: {df.shape}")
print(f"Filas: {df.shape[0]}, Columnas: {df.shape[1]}")
```

```python
# Ver las primeras 5 filas
print("=== PRIMERAS 5 FILAS ===")
print(df.head())

# Información general: tipos de datos y valores nulos
print("\n=== INFORMACIÓN GENERAL ===")
df.info()

# Estadísticas descriptivas de columnas numéricas
print("\n=== ESTADÍSTICAS DESCRIPTIVAS ===")
print(df.describe())

# Cuántos valores únicos tiene cada columna
print("\n=== VALORES ÚNICOS POR COLUMNA ===")
print(df.nunique())
```

```python
print("=== DISTRIBUCIÓN POR SEXO ===")
print(df["sexo"].value_counts())

print("\n=== DIAGNÓSTICO GLUCÉMICO ===")
print(df["diagnostico_glucemico"].value_counts())

print("\n=== CATEGORÍA DE PRESIÓN ARTERIAL ===")
print(df["categoria_TA"].value_counts())
```

```python
# Seleccionar una sola columna
glucosas = df["glucosa_ayuno"]
print(f"Tipo: {type(glucosas)}")
print(f"Promedio: {glucosas.mean():.1f} mg/dL")

# Seleccionar varias columnas
df_clinico = df[["id", "edad", "sexo", "imc", "glucosa_ayuno", "hba1c"]]
print(f"\nDataFrame reducido: {df_clinico.shape}")
print(df_clinico.head())

# Filtrar filas — pacientes diabéticos
diabeticos = df[df["diagnostico_glucemico"] == "Diabetes tipo 2"]
print(f"\nPacientes diabéticos: {len(diabeticos)}")

# Filtrar con dos condiciones — alto riesgo
alto_riesgo = df[(df["edad"] > 60) & (df["imc"] > 30)]
print(f"Alto riesgo cardiometabólico: {len(alto_riesgo)}")
print(alto_riesgo[["id", "edad", "imc", "diagnostico_glucemico"]])
```

```python
# Calcular índice glucosa-IMC
df["indice_glucosa_imc"] = df["glucosa_ayuno"] * df["imc"]

# Ver distribución del índice
print("=== ÍNDICE GLUCOSA-IMC ===")
print(f"Mínimo:  {df['indice_glucosa_imc'].min():.1f}")
print(f"Máximo:  {df['indice_glucosa_imc'].max():.1f}")
print(f"Promedio: {df['indice_glucosa_imc'].mean():.1f}")

# Clasificar en terciles de riesgo
df["riesgo_glucosa_imc"] = pd.cut(
    df["indice_glucosa_imc"],
    bins=3,
    labels=["Bajo", "Moderado", "Alto"]
)

print("\n=== DISTRIBUCIÓN DE RIESGO ===")
print(df["riesgo_glucosa_imc"].value_counts())

# Comparar con diagnóstico glucémico
print("\n=== RIESGO GLUCOSA-IMC × DIAGNÓSTICO ===")
print(pd.crosstab(df["riesgo_glucosa_imc"], df["diagnostico_glucemico"]))
```

```python
# Estadísticas por diagnóstico glucémico
resumen = df.groupby("diagnostico_glucemico")[["edad", "imc", "glucosa_ayuno", "hba1c", "indice_glucosa_imc"]].mean()
print("=== PROMEDIOS POR DIAGNÓSTICO ===")
print(resumen.round(1))

# Contar pacientes por sexo y diagnóstico
print("\n=== DISTRIBUCIÓN SEXO × DIAGNÓSTICO ===")
print(pd.crosstab(df["sexo"], df["diagnostico_glucemico"]))
```

```python
df.to_csv("../02_python_clinico/pacientes_simulados.csv", index=False)
print("CSV actualizado con indice_glucosa_imc y riesgo_glucosa_imc")
print(f"Columnas finales: {list(df.columns)}")
```
